### Solution

-The Differential semi-gradient TD(0) for estimating $\hat{q} \approx q_*$

--Input: a differentiable action-value function parameterization $\hat{q} : S \times A \times R^d \rightarrow R$

--Algorithm parameters: step sizes $\alpha$, $\beta > 0$

--Initialize value-function weights $W \in R^d$ arbitrarily (e.g., w = 0)

--Initialize average reward estimate $\bar{R} \in R$ arbitrarily (e.g., $bar{R} = 0$)

--**Initialize state S**

--Loop for each step:

   **Choose A as a function of $\hat{q}(S', . , W)$ (e.g., $\epsilon$-greedy)**

   **Take action A, observe R, S'**

   $\delta \leftarrow R - \bar{R} + \text{argmax} \; \bar{q}(S', A, W) - \bar{q}(S, A, W)$

   $\bar{R} \leftarrow \bar{R} + \beta \delta$

   $w \leftarrow w + \alpha \delta \nabla \bar{q}(S, A) $

   $S' \leftarrow S $
