### Exercise 4.1 
In Example 4.1, if $\pi$ is the equiprobable random policy, what is $q_{\pi}(11,down)$? What is $q_{\pi}(7,down)$?

### Answer:
Based on this equation:
$$q_{\pi}(s,a) = E[R_{t+1} + \gamma v_{\pi}(S_{t+1})|S_t = s, A_t = a]
	= \sum_{s, s’} p(s’, r|s,a)[r + \gamma v_{\pi}(s’)] 
$$

$p(11, r|T,down)=0$, $V^{\pi}(T) = 0$, so $q_{\pi}(11,down) = p(11, r|T,down) \times 1 + \gamma \times p(11, r|T,down) \times V^{\pi}(T) = 0 \gamma$

If $\gamma=1$, $q_{\pi}(11,down) = 0$ 

For the second case: 

$p(11, r|7,down)=-1$, $V^{\pi}(11) = -14$, so $q_{\pi}(11,down) = p(11, r|7,down) \times 1 + \gamma \times p(11, r|7,down) \times V^{\pi}(11) = -15 \gamma$

If $\gamma=1$, $q_{\pi}(7,down) = -15$

### Exercise 4.2 

In Example 4.1, suppose a new state 15 is added to the gridworld just below state 13, and its actions, left, up, right, and down, take the agent to states 12, 13, 14, and 15, respectively. Assume that the transitions from the original states are unchanged. What, then, is $v^{\pi}(15)$ for the equiprobable random policy? Now suppose the dynamics of state 13 are also changed, such that action down from state 13 takes the agent to the new state 15. What is $v^{\pi}(15)$ for the equiprobable random policy in this case?

$$ v_{k+1}(s) = \sum_a \pi(a|s) \sum_{s’, r} p(s’, r|s, a)[r + \gamma v_k (s’)] $$

What is $\pi(a|s)$ in this example?

### Solution:

$ V^{\pi}(15) = \sum_a \pi(a|15) \sum_{s’, r} p(s’, r|15, a)[r + \gamma v_k (s’)]
            = \frac{1}{4}(r(left, 15, 12) + \gamma V(12) + r(up, 15, 13) + \gamma V(13) \\
            + r(right, 15, 14) + \gamma V(14) + r(down, 15,15) + \gamma V(15)) $ \\
            
$ V^{\pi}(15) = \frac{1}{4}(-1 -22\gamma -1 -20\gamma -1 - 14\gamma -1 + \gamma V(15)) \\
        = -1 -14\gamma + \frac{\gamma}{4} V(15) $
        
$ V(15) = - \frac{4(1+14\gamma)}{4-\gamma} $, if let $\gamma = 1$,
